#Bar Chart

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
classes=["python","R","AI","ML","DS"]
class1_students=[30,40,27,56,61]
class2_students=[53,45,67,16,31]
class3_students=[49,74,41,38,75]

```

```python
plt.bar(classes,class1_students)
```

```python
#To change the orientation of graph
```

```python
plt.barh(classes,class1_students)
```

```python
#To change the width of the parameter
```

```python
plt.bar(classes,class1_students,width=0.4)
```

```python
#To change the alignment of bar by default it is centre ...now we are going to change the align at edge
```

```python
plt.bar(classes,class1_students,width=0.4,align="edge")
```

```python
#To change the color of the graph
```

```python
plt.bar(classes,class1_students,width=0.4,align="edge",color="y")
```

```python
#To change the colour of edge of graph
```

```python
plt.bar(classes,class1_students,width=0.4,align="edge",color="y",edgecolor="r")
```

```python
#To change the width of the line
```

```python
plt.bar(classes,class1_students,width=0.4,align="edge",color="y",edgecolor="r",linewidth=5)
```

```python
#To change the brightness of graph bar by using alpha parameter the range of alpha paremeter lies betweeen 0 to 1 abd it take float value
```

```python
plt.bar(classes,class1_students,width=0.4,align="edge",color="y",edgecolor="r",linewidth=5,alpha=0.4)
```

```python
#To change the linestyle of bar by using linestyle parameter
```

```python
plt.bar(classes,class1_students,width=0.4,align="edge",color="y",edgecolor="r",linewidth=3,alpha=0.4,linestyle="--")
```

```python
#To give the label to the graph
```

```python
plt.bar(classes,class1_students,width=0.4,align="edge",color="y",edgecolor="r",linewidth=3,alpha=0.4,linestyle="--",label="class1_students")
plt.legend()
```

```python
#To hide the graph we use visible parameter by default it is set to true
```

```python
plt.bar(classes,class1_students,width=0.4,align="edge",color="y",edgecolor="r",linewidth=3,alpha=0.4,linestyle="--",label="class1_students",visible=False)
plt.legend()
```

```python
plt.bar(classes,class1_students,width=0.4,align="edge",color="y",edgecolor="r",linewidth=3,alpha=0.4,linestyle="--",label="class1_students",visible=True)
plt.legend()
```

```python
#To channge the background of the graph
```

```python
style.use("ggplot")
plt.grid(color='k', linestyle='-.', linewidth=1)
plt.bar(classes,class1_students,width=0.4,align="edge",color="y",edgecolor="r",linewidth=3,alpha=0.4,linestyle="--",label="class1_students",visible=True)
plt.legend()
```

```python
#To change the size of figure
```

```python
style.use("ggplot")
plt.figure(figsize=(16,9))
plt.grid(color='k', linestyle='-.', linewidth=1)
plt.bar(classes,class1_students,width=0.4,align="edge",color="y",edgecolor="r",linewidth=3,alpha=0.4,linestyle="--",label="class1_students",visible=True)
plt.legend()
```

```python
#To make the above graph horizontal ...but for that we have to remove width prameter otherwise it can gives an error
```

```python
style.use("ggplot")
plt.figure(figsize=(16,9))
plt.grid(color='k', linestyle='-.', linewidth=1)
plt.barh(classes,class1_students,width=0.4,align="edge",color="y",edgecolor="r",linewidth=3,alpha=0.4,linestyle="--",label="class1_students",visible=True)
plt.legend()
```

```python
style.use("ggplot")
plt.figure(figsize=(16,9))
plt.grid(color='k', linestyle='-.', linewidth=1)
plt.barh(classes,class1_students,align="edge",color="y",edgecolor="r",linewidth=3,alpha=0.4,linestyle="--",label="class1_students",visible=True)
plt.legend()
```

```python
#To give the title,xlabel and ylabel
```

```python
plt.bar(classes,class1_students,width=0.4,align="edge",color="y",edgecolor="r",linewidth=3,alpha=0.4,linestyle="--",label="class1_students",visible=True)
plt.legend()
plt.title("classes/class1_students",fontsize=15)
plt.xlabel("Classes",fontsize=10)
plt.ylabel("Class1_students",fontsize=10)
plt.show()   #......necessary to use in othe IDLE except Jupyter-Notebook
```

```python
#To add more graphs
```

```python
plt.bar(classes,class1_students,width=0.4,color="b",label="class1_students")
plt.bar(classes,class2_students,width=0.4,color="g",label="class2_students")
plt.legend()
plt.title("classes/class_students",fontsize=15)
plt.xlabel("Classes",fontsize=10)
plt.ylabel("Classes_students",fontsize=10)
plt.show()   #......necessary to use in othe IDLE except Jupyter-Notebook
```

```python
#To avoid the overlapping of the graphs
```

```python
classes_index=np.arange((len(classes)))
width=0.2
plt.bar(classes_index,class1_students,width,color="b",label="class1_students")
plt.bar(classes_index + width,class2_students,width,color="g",label="class2_students")
plt.legend()
plt.title("classes/class_students",fontsize=15)
plt.xlabel("Classes",fontsize=10)
plt.ylabel("Classes_students",fontsize=10)
plt.show()   #......necessary to use in othe IDLE except Jupyter-Notebook
```

```python
#To replace the classes_index with their orginal values then we used xticks function
```

```python
classes_index=np.arange((len(classes)))
width=0.2
plt.bar(classes_index,class1_students,width,color="b",label="class1_students")
plt.bar(classes_index + width,class2_students,width,color="g",label="class2_students")
plt.legend()
plt.xticks(classes_index,classes)
plt.title("classes/class_students",fontsize=15)
plt.xlabel("Classes",fontsize=10)
plt.ylabel("Classes_students",fontsize=10)
plt.show()   #......necessary to use in othe IDLE except Jupyter-Notebook
```

```python
classes_index=np.arange((len(classes)))
width=0.2
plt.bar(classes_index,class1_students,width,color="b",label="class1_students")
plt.bar(classes_index + width,class2_students,width,color="g",label="class2_students")
plt.legend()
plt.xticks(classes_index + width,classes)
plt.title("classes/class_students",fontsize=15)
plt.xlabel("Classes",fontsize=10)
plt.ylabel("Classes_students",fontsize=10)
plt.show()   #......necessary to use in othe IDLE except Jupyter-Notebook
```

```python
#To change the rotation of classes values we use rotation function in xtick argument...rotation(angle in degree)
```

```python
classes_index=np.arange((len(classes)))
width=0.2
plt.bar(classes_index,class1_students,width,color="b",label="class1_students")
plt.bar(classes_index + width,class2_students,width,color="g",label="class2_students")
plt.legend()
plt.xticks(classes_index + width,classes,rotation(90))
plt.title("classes/class1_students",fontsize=15)
plt.xlabel("Classes",fontsize=10)
plt.ylabel("Classes_students",fontsize=10)
plt.show()   #......necessary to use in othe IDLE except Jupyter-Notebook
```
