#Histogram

```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
ml_student_age=np.random.randint(18,45,(100))
py_student_age=np.random.randint(38,73,(100))
```

```python
print(ml_student_age)
print(py_student_age)
```

```python
plt.hist(ml_student_age)
plt.title("ml_student_age histogram")
plt.xlabel("Student age Category")
plt.ylabel("Number of student age")
plt.show()
```

```python
#To give the proper axis point to x-axis we used bins function...
```

```python
bins=[]
for i in range(18,75,2):
    bins.append(i)
print(bins) 
plt.hist(ml_student_age,bins)
plt.title("ml_student_age histogram")
plt.xlabel("Student age Category")
plt.ylabel("Number of student age")
plt.show()
```

```python
#To adjust the bar width use rwidth function bydefault its space is 1 means 100%
```

```python
bins=[]
for i in range(18,75,2):
    bins.append(i)
print(bins) 
plt.hist(ml_student_age,bins,rwidth=0.8)
plt.title("ml_student_age histogram")
plt.xlabel("Student age Category")
plt.ylabel("Number of student age")
plt.show()
```

```python
#for histtype we can pass parameter as {'bar','barstacked','step','stepfilled'}
```

```python
bins=[]
for i in range(18,75,2):
    bins.append(i)
print(bins) 
plt.hist(ml_student_age,bins,rwidth=0.8,histtype='step')
plt.title("ml_student_age histogram")
plt.xlabel("Student age Category")
plt.ylabel("Number of student age")
plt.show()
```

```python
bins=[]
for i in range(18,75,2):
    bins.append(i)
print(bins) 
plt.hist(ml_student_age,bins,rwidth=0.8,histtype='stepfilled')
plt.title("ml_student_age histogram")
plt.xlabel("Student age Category")
plt.ylabel("Number of student age")
plt.show()
```

```python
bins=[]
for i in range(18,75,2):
    bins.append(i)
print(bins) 
plt.hist(ml_student_age,bins,rwidth=0.8,histtype='bar')
plt.title("ml_student_age histogram")
plt.xlabel("Student age Category")
plt.ylabel("Number of student age")
plt.show()
```

```python
bins=[]
for i in range(18,75,2):
    bins.append(i)
print(bins) 
plt.hist(ml_student_age,bins,rwidth=0.8,histtype='barstacked')
plt.title("ml_student_age histogram")
plt.xlabel("Student age Category")
plt.ylabel("Number of student age")
plt.show()
```

```python
#To change the orientation of graph {horizontal,vertical}
```

```python
bins=[]
for i in range(18,75,2):
    bins.append(i)
print(bins) 
plt.hist(ml_student_age,bins,rwidth=0.8,histtype='barstacked',orientation="horizontal")
plt.title("ml_student_age histogram")
plt.xlabel("Student age Category")
plt.ylabel("Number of student age")
plt.show()
```

```python
#To change the color of graph
```

```python
bins=[]
for i in range(18,75,2):
    bins.append(i)
print(bins) 
plt.hist(ml_student_age,bins,rwidth=0.8,histtype='barstacked',orientation="horizontal",color="m")
plt.title("ml_student_age histogram")
plt.xlabel("Student age Category")
plt.ylabel("Number of student age")
plt.show()
```

```python
#To give the label to graph
```

```python
bins=[]
for i in range(18,75,2):
    bins.append(i)
print(bins) 
plt.hist(ml_student_age,bins,rwidth=0.8,histtype='barstacked',orientation="horizontal",color="m",label="ML student")
plt.title("ml_student_age histogram")
plt.xlabel("Student age Category")
plt.ylabel("Number of student age")
plt.legend()
plt.show()
```

```python
#to add py_student age
```

```python
bins=[]
for i in range(18,75,2):
    bins.append(i)
print(bins) 
plt.hist(ml_student_age,bins,rwidth=0.8,histtype='barstacked',orientation="horizontal",color="m",label="ML student")
plt.hist(py_student_age,bins,rwidth=0.8,histtype='barstacked',orientation="vertical",color="b",label="py student")
plt.title("ml_student_age histogram")
plt.xlabel("Student age Category")
plt.ylabel("Number of student age")
plt.legend()
plt.show()
```

```python
bins=[]
for i in range(18,75,2):
    bins.append(i)
print(bins) 
plt.hist([ml_student_age,py_student_age],bins,rwidth=0.8,histtype='barstacked',orientation="vertical",color=["b","r"],label=["ML student","py student"])
plt.title("ml_student_age histogram")
plt.xlabel("Student age Category")
plt.ylabel("Number of student age")
plt.legend()
plt.show()
```

```python
#To change the size of the Histogram
```

```python
bins=[]
for i in range(18,75,2):
    bins.append(i)
print(bins) 
plt.figure(figsize=(20,30))
plt.hist([ml_student_age,py_student_age],bins,rwidth=0.8,histtype='barstacked',orientation="vertical",color=["b","r"],label=["ML student","py student"])
plt.title("ml_student_age histogram")
plt.xlabel("Student age Category")
plt.ylabel("Number of student age")
plt.legend()
plt.show()
```

```python
from matplotlib import style
style.use("ggplot")
bins=[]
for i in range(18,75,2):
    bins.append(i)
print(bins) 
plt.figure(figsize=(16,9))
plt.hist([ml_student_age,py_student_age],bins,rwidth=0.8,histtype='barstacked',orientation="vertical",color=["b","r"],label=["ML student","py student"])
plt.title("ml_student_age histogram")
plt.xlabel("Student age Category")
plt.ylabel("Number of student age")
plt.legend()
plt.show()
```

```python
from matplotlib import style
style.use("ggplot")
bins=[]
for i in range(18,75,2):
    bins.append(i)
print(bins) 
plt.figure(figsize=(16,9))
plt.hist([ml_student_age,py_student_age],bins,rwidth=0.8,histtype='barstacked',orientation="vertical",color=["b","r"],label=["ML student","py student"])
plt.title("ml_student_age histogram")
plt.xlabel("Student age Category")
plt.ylabel("Number of student age")
plt.legend()
plt.grid(color='k', linestyle='--', linewidth=1)
plt.show()
```
