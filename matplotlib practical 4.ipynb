```python
df=pd.read_csv('C:\\Users\\AS\\Downloads\\archive\\googleplaystore_user_reviews.csv',nrows=1000)
df
```

```python
df.fillna(0)
```

```python
x=df["Sentiment_Polarity"]
y=df["Sentiment_Subjectivity"]
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x,y)
plt.title("Google Play Store Sentiment_Polarity v/s Sentiment_Subjectivity graph",fontsize=20)
plt.xlabel("Sentiment_Polarity",fontsize=10)
plt.ylabel("Sentiment_Subjectivity",fontsize=10)
```

```python
#To change the color of scatter
```

```python
plt.scatter(x,y,color='r')
plt.title("Google Play Store Sentiment_Polarity v/s Sentiment_Subjectivity graph",fontsize=20)
plt.xlabel("Sentiment_Polarity",fontsize=10)
plt.ylabel("Sentiment_Subjectivity",fontsize=10)
```

```python
#To adjust the size of scatter plot graph
```

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y,color='r')
plt.title("Google Play Store Sentiment_Polarity v/s Sentiment_Subjectivity graph",fontsize=20)
plt.xlabel("Sentiment_Polarity",fontsize=10)
plt.ylabel("Sentiment_Subjectivity",fontsize=10)
```

```python
#To change the shape of scatter
```

```python
plt.scatter(x,y,color='r',marker='*')
plt.title("Google Play Store Sentiment_Polarity v/s Sentiment_Subjectivity graph",fontsize=20)
plt.xlabel("Sentiment_Polarity",fontsize=10)
plt.ylabel("Sentiment_Subjectivity",fontsize=10)
```

```python
plt.scatter(x,y,color='r',marker='+')
plt.title("Google Play Store Sentiment_Polarity v/s Sentiment_Subjectivity graph",fontsize=20)
plt.xlabel("Sentiment_Polarity",fontsize=10)
plt.ylabel("Sentiment_Subjectivity",fontsize=10)
```

```python
#To increase the size of  the parameter
```

```python
plt.scatter(x,y,color='r',marker='+',s=100)
plt.title("Google Play Store Sentiment_Polarity v/s Sentiment_Subjectivity graph",fontsize=20)
plt.xlabel("Sentiment_Polarity",fontsize=10)
plt.ylabel("Sentiment_Subjectivity",fontsize=10)
```

```python
#Alpha function
```

```python
plt.scatter(x,y,color='r',marker='+',s=100,alpha=0.3)
plt.title("Google Play Store Sentiment_Polarity v/s Sentiment_Subjectivity graph",fontsize=20)
plt.xlabel("Sentiment_Polarity",fontsize=10)
plt.ylabel("Sentiment_Subjectivity",fontsize=10)
```

```python
#To change the width of scatter
```

```python
plt.scatter(x,y,color='r',marker='+',s=100,alpha=0.3,linewidth=5)
plt.title("Google Play Store Sentiment_Polarity v/s Sentiment_Subjectivity graph",fontsize=20)
plt.xlabel("Sentiment_Polarity",fontsize=10)
plt.ylabel("Sentiment_Subjectivity",fontsize=10)
```

```python
#To change the edge colour
```

```python
plt.scatter(x,y,color='r',marker='*',s=100,alpha=0.3,linewidth=5,edgecolors='y')
plt.title("Google Play Store Sentiment_Polarity v/s Sentiment_Subjectivity graph",fontsize=20)
plt.xlabel("Sentiment_Polarity",fontsize=10)
plt.ylabel("Sentiment_Subjectivity",fontsize=10)
```

```python
#multiple scatters
```

```python
z=df["Sentiment"]
plt.scatter(x,y,color='r',marker='*',s=100,alpha=0.3,linewidth=5,edgecolors='y')
plt.scatter(x,z,color='g',marker='<',s=100,alpha=0.3,linewidth=5,edgecolors='k')
plt.title("Google Play Store Sentiment_Polarity v/s Sentiment_Subjectivity graph",fontsize=20)
plt.xlabel("Sentiment_Polarity",fontsize=10)
plt.ylabel("Sentiment_Subjectivity",fontsize=10)
```
