```python
#To show the values in the form of percentage
```

```python
classes=["Python","R","Machine Learning","Artificial Intelligence","Data Science"]
class_student=[45,15,35,25,30]
explode=[0.03,0,0.2,0,0]
colors=["c","b","r","y","g"]
plt.pie(class_student,labels=classes,explode=explode,colors=colors,autopct="%0.2f%%")
plt.show()
```

```python
#To manage the size of the texts..use textprops function
```

```python
classes=["Python","R","Machine Learning","Artificial Intelligence","Data Science"]
class_student=[45,15,35,25,30]
explode=[0.03,0,0.2,0,0]
colors=["c","b","r","y","g"]
textprops={"fontsize":20}
plt.pie(class_student,labels=classes,explode=explode,colors=colors,autopct="%0.2f%%",shadow=True,startangle=270,textprops=textprops)
plt.show()
```

```python
#wedgeprops
```

```python
classes=["Python","R","Machine Learning","Artificial Intelligence","Data Science"]
class_student=[45,15,35,25,30]
explode=[0.03,0,0.2,0,0]
colors=["c","b","r","y","g"]
textprops={"fontsize":20}
wedgeprops={"linewidth":4,'width':1,"edgecolor":"k"}
plt.pie(class_student,labels=classes,explode=explode,colors=colors,autopct="%0.2f%%",shadow=True,startangle=270,textprops=textprops,wedgeprops=wedgeprops)
plt.show()
```

```python
classes=["Python","R","Machine Learning","Artificial Intelligence","Data Science"]
class_student=[45,15,35,25,30]
explode=[0.03,0,0.2,0,0]
colors=["c","b","r","y","g"]
textprops={"fontsize":20}
wedgeprops={"linewidth":4,'width':2,"edgecolor":"k"}
plt.pie(class_student,labels=classes,explode=explode,colors=colors,autopct="%0.2f%%",shadow=True,startangle=270,textprops=textprops,wedgeprops=wedgeprops)
plt.show()
```

```python
#center(x,y)...where to plot pie chart in graph
```

```python
classes=["Python","R","Machine Learning","Artificial Intelligence","Data Science"]
class_student=[45,15,35,25,30]
explode=[0.03,0,0.2,0,0]
colors=["c","b","r","y","g"]
textprops={"fontsize":9}
wedgeprops={"linewidth":4,'width':1,"edgecolor":"k"}
center=(2,7)
plt.pie(class_student,labels=classes,explode=explode,colors=colors,autopct="%0.2f%%",shadow=True,startangle=270,textprops=textprops,wedgeprops=wedgeprops,center=center)
plt.show()
```

```python
#frame...by default it is false
```

```python
classes=["Python","R","Machine Learning","Artificial Intelligence","Data Science"]
class_student=[45,15,35,25,30]
explode=[0.03,0,0.2,0,0]
colors=["c","b","r","y","g"]
textprops={"fontsize":9}
wedgeprops={"linewidth":4,'width':1,"edgecolor":"k"}
center=(2,7)
plt.pie(class_student,labels=classes,explode=explode,colors=colors,autopct="%0.2f%%",shadow=True,startangle=270,textprops=textprops,wedgeprops=wedgeprops,center=center,frame=True)
plt.show()
```

```python
#To rotate the labels
```

```python
classes=["Python","R","Machine Learning","Artificial Intelligence","Data Science"]
class_student=[45,15,35,25,30]
explode=[0.03,0,0.2,0,0]
colors=["c","b","r","y","g"]
textprops={"fontsize":9}
wedgeprops={"linewidth":4,'width':1,"edgecolor":"k"}
center=(2,7)
plt.pie(class_student,labels=classes,explode=explode,colors=colors,autopct="%0.2f%%",shadow=True,startangle=270,textprops=textprops,wedgeprops=wedgeprops,center=center,frame=True,rotatelabels=True)
plt.show()
```
